```python
import os
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
base_dir = Path("../data/background")
csv_list = sorted(list(base_dir.glob("*.csv")))
```

```python
df_list = [pd.read_csv(file) for file in csv_list]
```

```python
fig, ax = plt.subplots(5,7)
for df, ax in zip(df_list,fig.get_axes()):
    #df['BIRTHRGN'].plot(kind="hist",ax=ax,bins=12,title=str(df["CNTRYID"]))
    category_counts = df['BIRTHRGN'].value_counts()

    plt.bar(category_counts.index, category_counts.values)

    for tick in ax.get_xticklabels():
        tick.set_rotation(45)
fig.tight_layout()
plt.show()
```
